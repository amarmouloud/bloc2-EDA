2.6 L'objectif principal des participants

3.3 Les objectifs des participants et les personnes rencontrées

```python
## Nous allons rechercher l'objectif principal entre les personnes et des personnes qu'elles sont rencontrées.

goal_person = dataset[['iid', 'goal']].drop_duplicates(keep='first')
```

```python
len(goal_person.iid.unique())
```

```python
goal_person.shape
```

```python
goals = dataset.merge(goal_person, how='inner', left_on='pid', right_on='iid', suffixes=('','_o'))
goals_match = goals[goals['match'] == 1]
goals_match = goals_match[['goal','goal_o']]
goals_not_match = goals[goals['match'] == 0]
goals_not_match = goals_not_match[['goal','goal_o']]
```
